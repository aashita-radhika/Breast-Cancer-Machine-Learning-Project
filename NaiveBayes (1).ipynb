```python
# Importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score,ConfusionMatrixDisplay
```

```python
# Read the data from 'NewData.csv' into a pandas DataFrame
df = pd.read_csv('NewData.csv')
```

```python
# Reviewing dataset
df
```

```python
# Extract all columns except the first one and store them in X as a NumPy array
X = df.iloc[:, 1:].values

# Extract the first column and store it in y as a NumPy array
y = df.iloc[:, 0].values
```

```python
# we split the data into training and testing data 
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2,random_state=2)
```

```python
# Scatterplot for the training data
plt.scatter(X_train[:,0], X_train[:,1], c = y_train, s = 50)
```

```python
from sklearn.naive_bayes import GaussianNB

# Using inbuilt Gaussian Classifier
model1 = GaussianNB()

# Fitting training data into the model
model1.fit(X_train,y_train)

# Predicting target values for testing data
y_pred1 = model1.predict(X_test)
```

```python
accuray = accuracy_score(y_pred1, y_test)
f1 = f1_score(y_pred1, y_test, average="weighted")
recall = recall_score(y_pred1, y_test)
precision=precision_score(y_pred1, y_test)

print("Accuracy:", accuray*100)
print("F1 Score:", f1*100)
print("Recall Score:", recall*100)
print("Precision Score:",precision*100)
```

```python
from sklearn.naive_bayes import MultinomialNB

# Using inbuilt Multinomial Classifier
model2 = MultinomialNB()

# Fitting training data into the model
model2.fit(X_train,y_train)

# Predicting target values for testing data
y_pred2 = model2.predict(X_test)

# Prdicting accuracy score,f1score,recall score, precision score

accuray = accuracy_score(y_pred2, y_test)
f1 = f1_score(y_pred2, y_test, average="weighted")
recall = recall_score(y_pred2, y_test)
precision=precision_score(y_pred2, y_test)

print("Accuracy:", accuray*100)
print("F1 Score:", f1*100)
print("Recall Score:", recall*100)
print("Precision Score:",precision*100)
```
